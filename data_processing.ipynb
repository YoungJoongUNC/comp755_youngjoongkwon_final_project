```python
def count_frames_manual(video):
    # initialize the total number of frames read
    total = 0
    
    # loop over the frames of the video
    while True:
        # grab the current frame
        (grabbed, frame) = video.read()
        
        # check to see if we have reached the end of the video
        if not grabbed:
            break
        # increment the total number of frames read
        total += 1
        
    # return the total number of frames in the video file
    return total
```

```python
def count_frames(path, override=False):
    # grab a pointer to the video file and initialize the total number of frames read
    video = cv2.VideoCapture(path)
    total = 0
    fps = 0
    
    # if the override flag is passed in, revert to the manual method of counting frames
    if override:
        total = count_frames_manual(video)
    # otherwise, let's try the fast way first
    else:
        total = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
        fps = int(video.get(cv2.CAP_PROP_FPS))
    
    # release the video file pointer
    video.release()
    
    # return the total number of frames in the video
    return (total, fps)
```

```python
def get_video_running_time(path):
    
    total_number_of_frame,fps = count_frames(path)
    
    # return total video running time in second
    return int(total_number_of_frame/fps)
```

```python
def process_a_video_into_frames(raw_video_path, video_index_base, num_of_classes,total_frames_per_big_segment):
    
    # this function makes small videos out of one input video
    # this function return total number of generated small video segments
    # name of each small video : video index_frame number 
    # ex) frame#11 of video#8 would be 8_11
    
    video = cv2.VideoCapture(raw_video_path)
    
    raw_frame_index = -1 # frame index in raw video
    big_segment_index = -1
    
    video_index_offset = 0 # final vide_index would be video_index_base + video_index_offset
    video_frame_index = 0
    
    total_number_of_frame, fps = count_frames(raw_video_path)
    expected_total_number_of_video = int(total_number_of_frame/total_frames_per_big_segment)*num_of_classes
    total_number_of_generated_segments = 0
    
    
    while True:
        # grab the current frame
        (grabbed, frame) = video.read()
        
        # check to see if we have reached the end of the video
        if not grabbed:
            break
            
        # increment the frame index
        raw_frame_index += 1
        
        if (raw_frame_index%total_frames_per_big_segment)==0:
            big_segment_index+=1
        
        
        
        
        video_index_offset = num_of_classes*big_segment_index+(raw_frame_index - (total_frames_per_big_segment*big_segment_index))%num_of_classes
        
        if video_index_offset >= expected_total_number_of_video :
            continue
        
        total_number_of_generated_segments = video_index_offset
        
        video_frame_index = (raw_frame_index - (total_frames_per_big_segment*big_segment_index))//num_of_classes 
        output_file_name = str(video_index_base+video_index_offset)+"_"+str(video_frame_index)+".jpg" 
        cv2.imwrite(os.path.join(SMALL_VIDEO_OUTPUT_PATH, output_file_name),frame)
        
        
            
        
    video.release()
    return total_number_of_generated_segments+1
```

```python
def process_all_videos(videos_dir_path):
    
    # process MIT Fabric Dataset into small video segments
    # outputs (video index , label)
    
    ############### argument list ##################################
    video_index_base = 0
    num_of_classes = 9
    total_frames_per_big_segment = 135 # 135 frames per big segment
    ################################################################
    ground_truth = [] # (video index , material index)
    
    video_index = 0
    
    for file in os.listdir(videos_dir_path):
        
        raw_video_path = os.path.join(videos_dir_path, file)
        
        start_video_index = video_index_base
        
        video_index_base += process_a_video_into_frames(raw_video_path, video_index_base, num_of_classes, total_frames_per_big_segment)
        
        end_video_index = video_index_base
        
        for idx in range(start_video_index, end_video_index):
            ground_truth.append((idx, video_index//3))
        
        video_index += 1
    
    return ground_truth
```

```python
videos_dir_path = 'D:\MIT Fabric Dataset'
ground_truth = process_all_videos(videos_dir_path)
for gt in ground_truth:
    print(str(gt[0])+" "+str(gt[1]))
```

```python
def make_mat_videoindex_list(groundtruth_list):
    
    # make 2D list mat[i] : [ list of video index ]
    mat_videoindex_list = []
    # initialize the list
    for i in range(0,30):
        mat_videoindex_list.append([])
    
    for gt in groundtruth_list:
        mat_videoindex_list[gt[1]].append(gt)
    
    return mat_videoindex_list
```

```python
mat_videoindex_list = make_mat_videoindex_list(ground_truth)
for i in range(0,30):
    for j in range(len(mat_videoindex_list[i])):
        elem = mat_videoindex_list[i][j]
        print("video index: "+str(elem[0])+" mat label: "+str(elem[1]))
```

```python
def split_train_valid_test_dataset(mat_videoindex_list):
    
    # this function returns dataset (video index, material label) array (training/validation/testing video index)
    
    # initialize dataset array
    # dataset[0] = training dataset
    # dataset[1] = validation dataset
    # dataset[2] = testing dataset
    
    dataset=[]
    for i in range(3):
        dataset.append([])
        
    val_percent = 0.05
    test_percent = 0.25
    
    for i in range(30):
        #print("mat"+str(i))
        total_num = len(mat_videoindex_list[i])
        #print(total_num)
        val_num = int(val_percent*total_num)
        #print(val_num)
        test_num = int(test_percent*total_num)
        #print(test_num)
        permed_index = np.random.permutation(total_num)
        
        for j in range(total_num):
            
            if j < val_num:
                dataset[1].append(mat_videoindex_list[i][permed_index[j]])
            elif val_num <= j and j < val_num+test_num:
                dataset[2].append(mat_videoindex_list[i][permed_index[j]])
            else:
                dataset[0].append(mat_videoindex_list[i][permed_index[j]])
    
    return dataset
            
```

```python
dataset=split_train_valid_test_dataset(mat_videoindex_list)
```
